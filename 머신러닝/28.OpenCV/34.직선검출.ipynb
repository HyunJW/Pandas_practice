```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('c:/data/images/road.jpg')
src = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(src)
plt.axis('off')
plt.show()
```

```python
# -    
#                
import numpy as np
dst = img.copy()
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
#  5000, 1500,   
canny = cv2.Canny(gray, 5000, 1500, apertureSize=5, L2gradient=True)
# cv2.HoughLines(, , , ,  ,  ,  ,  )
lines = cv2.HoughLines(canny, 0.8, np.pi/180, 150, srn=100, stn=200, min_theta=0, max_theta=np.pi)
for i in lines:
    rho, theta = i[0][0], i[0][1]
    a, b = np.cos(theta), np.sin(theta)
    x0, y0 = a*rho, b*rho
    scale = src.shape[0] + src.shape[1]
    x1 = int(x0 + scale*(-b))
    y1 = int(y0 + scale*a)
    x2 = int(x0 - scale*(-b))
    y2 = int(y0 - scale*a)
    cv2.line(dst, (x1,y1), (x2,y2), (0,0,255), 2)
    cv2.circle(dst, (int(x0),int(y0)), 3, (255,0,0), 5, cv2.FILLED)
```

```python
src = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(src)
plt.title('image1')
plt.axis("off")
plt.show()
```

```python
# -    (Progressive Probabilistic Hough Transform)
#        
dst = img.copy()
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(gray, 5000, 1500, apertureSize=5, L2gradient=True)
lines = cv2.HoughLinesP(canny, 0.8, np.pi/180, 90, minLineLength=10, maxLineGap=100)
for i in lines:
    cv2.line(dst, (int(i[0][0]),int(i[0][1])), (int(i[0][2]),int(i[0][3])), (0,0,255), 2)
```

```python
src = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(src)
plt.title('image2')
plt.axis("off")
plt.show()
```
