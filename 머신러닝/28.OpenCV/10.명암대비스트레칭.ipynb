```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("c:/data/images/winter.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.axis("off")
plt.show()
```

```python
out = gray.copy()
height, width = gray.shape
height, width
```

```python
#명도값 = ((원래값 - 최저명도값) * 255) / (최고명도값 - 최저명도값)
# 밝게 하는 경우
low = 75
high = 255
for i in range(height):
    for j in range(width):
        out[i][j] = ((gray[i][j] - low) * 255 / (high - low))
```

```python
#비교
plt.figure(figsize=(20,6))
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(out, cmap='gray')
plt.axis("off")
plt.show()
```

```python
#픽셀에 대한 히스토그램(밝기에 해당하는 명도값): 오른쪽으로 치우쳐 있으면 명암대비가 높은 이미지
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.hist(gray.ravel(), 256, [0, 256])
plt.subplot(1, 2, 2)
plt.hist(out.ravel(), 256, [0, 256])
plt.show()
```

```python
# 어둡게 하는 경우
low = 0
high = 200
for i in range(height):
    for j in range(width):
        out[i][j] = ((gray[i][j] - low) * 255 / (high - low))
```

```python
#비교
plt.figure(figsize=(20,6))
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(out, cmap='gray')
plt.axis("off")
plt.show()
```

```python
#픽셀에 대한 히스토그램: 왼쪽으로 치우쳐 있으면 명암대비가 낮은 이미지
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.hist(gray.ravel(), 256, [0, 256])
plt.subplot(1, 2, 2)
plt.hist(out.ravel(), 256, [0, 256])
plt.show()
```
