```python
#히스토그램 평탄화: 히스토그램 특정 영역에 집중되어 있는 부분을 평탄하게 만들어주는 작업
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("c:/data/images/apples.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.axis("off")
plt.show()
```

```python
import numpy as np
histogram, bin = np.histogram(img.ravel(), 256, [0, 256])
cumsum = histogram.cumsum()
```

```python
#LUT(Look Up Table, 룩업 테이블): A = B의 집합, A를 입력하면 B가 리턴
LUT = np.uint8((cumsum - cumsum.min()) * 255 / (cumsum.max() - cumsum.min()))
equ = LUT[gray]
hist = cv2.equalizeHist(gray) # gray이미지에만 적용 가능
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1, 3, 1)
plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(equ, cmap='gray')
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(hist, cmap='gray')
plt.axis("off")
plt.show()
```

```python
#픽셀에 대한 히스토그램
plt.figure(figsize=(20,6))
plt.subplot(1, 3, 1)
plt.hist(img.ravel(), 256, [0, 256]) # 왼쪽으로 치우친 히스토그램
plt.axis("off")
plt.subplot(1, 3, 2)
plt.hist(equ.ravel(), 256, [0, 256]) # 직접 계산
plt.axis("off")
plt.subplot(1, 3, 3)
plt.hist(hist.ravel(), 256, [0, 256]) # 함수로 계산
plt.axis("off")
plt.show()
```
