```python
# :         
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("c:/data/images/apples.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.axis("off")
plt.show()
```

```python
import numpy as np
histogram, bin = np.histogram(img.ravel(), 256, [0, 256])
cumsum = histogram.cumsum()
```

```python
#LUT(Look Up Table,  ): A = B , A  B 
LUT = np.uint8((cumsum - cumsum.min()) * 255 / (cumsum.max() - cumsum.min()))
equ = LUT[gray]
hist = cv2.equalizeHist(gray) # gray  
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1, 3, 1)
plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(equ, cmap='gray')
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(hist, cmap='gray')
plt.axis("off")
plt.show()
```
