```python
#영상 이진화: 특정 경계값을 기준으로 0, 255로 만드는 작업
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('c:/data/images/apples.jpg')
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGBA)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
#픽셀값 127을 기준으로 0, 255로 나누기
ret, dst = cv2.threshold(image_rgb, 127, 255, cv2.THRESH_BINARY)
print(ret) # 경계값
print(dst) # 출력값
```

```python
plt.imshow(dst, cmap='gray')
plt.axis('off')
plt.show()
```
