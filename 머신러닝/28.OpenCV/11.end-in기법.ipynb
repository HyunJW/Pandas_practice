```python
#end-in :  +             
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("c:/data/images/apples.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.axis("off")
plt.show()
```

```python
out = gray.copy()
height, width = gray.shape
height, width
```

```python
#   
MIN = 30
MAX = 255
```

```python
#  (10 0, 240 255)
low = 10
high = 240
```

```python
for i in range(height):
    for j in range(width):
        if gray[i][j] <= low:
            out[i][j] = 0
        elif gray[i][j] >= high:
            out[i][j] = 255
        else:
            out[i][j] = ((gray[i][j] - MIN) * 255) / (MAX - MIN)
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1, 2, 1)
plt.imshow(gray,cmap='gray')
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(out,cmap='gray')
plt.axis("off")
plt.show()
```

```python
#  
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.hist(gray.ravel(), 256, [0, 256])
plt.subplot(1, 2, 2)
plt.hist(out.ravel(), 256, [0, 256])
plt.show()
```
