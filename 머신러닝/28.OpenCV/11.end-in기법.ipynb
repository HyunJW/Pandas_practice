```python
#end-in 기법: 스트레칭기법 + 양 끝 일정 부분을 잘라내어 이미지의 명암 대비가 더 뚜렷해 보이게 하는 기법
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("c:/data/images/apples.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.axis("off")
plt.show()
```

```python
out = gray.copy()
height, width = gray.shape
height, width
```

```python
#이미지의 최솟값과 최댓값 지정
MIN = 30
MAX = 255
```

```python
#하한값과 상한값 지정(10이하는 0, 240이상은 255)
low = 10
high = 240
```

```python
for i in range(height):
    for j in range(width):
        if gray[i][j] <= low:
            out[i][j] = 0
        elif gray[i][j] >= high:
            out[i][j] = 255
        else:
            out[i][j] = ((gray[i][j] - MIN) * 255) / (MAX - MIN)
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1, 2, 1)
plt.imshow(gray,cmap='gray')
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(out,cmap='gray')
plt.axis("off")
plt.show()
```

```python
#픽셀에 대한 히스토그램
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.hist(gray.ravel(), 256, [0, 256])
plt.subplot(1, 2, 2)
plt.hist(out.ravel(), 256, [0, 256])
plt.show()
```
