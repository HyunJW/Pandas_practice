```python
import pandas as pd
df = pd.read_csv('c:/data/time/nile.csv')
df
```

```python
df.plot() # 불안정한 시계열
```

```python
#1차 차분 데이터
import matplotlib.pyplot as plt
diff1 = df['x'].diff(periods=1).iloc[1:]
diff1.plot()
plt.show()
```

```python
#로그 후 차분한 자료가 안정적인 시계열인지 확인
from statsmodels.tsa.stattools import adfuller
import numpy as np
diff1 = np.log(df['x']).diff(periods=1).iloc[1:]
result = adfuller(diff1)
print(result)
print('ADF Statistic: %f' %result[0])
print('p-value: %f' %result[1])
```

```python
#가장 적절한 모형
from pmdarima.arima import auto_arima
model = auto_arima(df, suppress_warnings=True)
model
```

```python
#학습용, 검증용 구분
n = int(df.shape[0] * 0.8)
train = df[:n]
test = df[n:]
print(len(df['x']))
print(train['x'].values)
print(test['x'].values)
```

```python
#가장 적절한 모형 생성
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(train, order=(1,1,1)).fit()
model.plot_diagnostics(figsize=(16,8))
plt.show()
```

```python
#예측값
predictions = model.predict(start=1, end=len(test), dynamic=False)
predictions.index = len(train) + (predictions.index) - 1
predictions
```

```python
#시각화
plt.plot(df, label='Real')
plt.plot(predictions, label='Predict')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
#오차
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test, predictions)
rmse = np.sqrt(mse)
rmse
```
