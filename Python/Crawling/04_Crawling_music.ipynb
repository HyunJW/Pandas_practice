```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# 멜론차트 크롤링

```python
url = 'https://www.melon.com/chart/index.htm'
req = requests.get(url)
req.text
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
req = requests.get(url, headers=header)
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
rank = soup.select('.rank')[100].text
rank
```

```python
song = soup.select('.ellipsis.rank01 a')[0].text
song
```

```python
artist = soup.select('.ellipsis.rank02 a')[0].text
artist
```

```python
album = soup.select('.ellipsis.rank03 a')[2].text
album
```

```python
li = []
for i in range(100):
    rank = soup.select('.rank')[i + 1].text
    title = soup.select('.ellipsis.rank01 a')[i].text
    artist = soup.select('.ellipsis.rank02 a')[i].text
    album = soup.select('.ellipsis.rank03 a')[i].text
    li.append({'순위' : rank, '제목' : title, '가수' : artist, '앨범' : album})
li
```

```python
df = pd.DataFrame(li)
df
```

```python
df.to_csv('멜론차트 top100.csv', index=False)
```

# 지니차트 크롤링

```python
genie = []
for i in range(4):
    url = f'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230111&hh=16&rtm=Y&pg={i+1}'
    req = requests.get(url)
    req.text
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
    req = requests.get(url, headers=header)
    soup = BeautifulSoup(req.text, 'html.parser')
    
    for li in soup.select('tr.list'):
        try:
            rank = li.select_one('.number').text.split('\n')[0]
        except:
            pass
        title = li.select('.info > a')[0].text.split('\n')[-1].lstrip()
        artist = li.select('.info > a')[1].text.split('\n')[-1].lstrip()
        album = li.select('.info > a')[2].text.split('\n')[-1].lstrip()
        genie.append({'순위' : rank, '제목' : title, '가수' : artist, '앨범' : album})
```

```python
df = pd.DataFrame(genie)
df
```

```python
df.to_csv('지니차트 top200.csv', index=False)
```
