# 식신 크롤링

```python
site = 'https://www.siksinhot.com/'
url = f'{site}/search?keywords={quote("신논현역")}'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

## 필요한 데이터 찾기

```python
lis = soup.select_one('.localFood_list').find_all('li')
len(lis)
```

```python
lis = soup.select('.localFood_list > li')
len(lis)
```

```python
li = lis[0]
li
```

```python
store = li.select_one('.textBox > h2').get_text()
store
```

```python
# float(), int()로 변환하고자 할때는 반드시 예외처리를 해줘야 함.
# 점수가 없는 경우가 있기 때문에.
try:
    score = float(li.select_one('.score').get_text())
except:
    score = 0.0
```

```python
score
```

```python
place = li.select('.cate > a')[0].text
```

```python
place
```

```python
menu = li.select('.cate > a')[1].text
```

```python
menu
```

```python
line = []
for li in lis:
    store = li.select_one('.textBox > h2').get_text()
    try:
        score = float(li.select_one('.score').get_text())
    except:
        score = 0.0
    place = li.select('.cate > a')[0].text
    menu = li.select('.cate > a')[1].text
    line.append({'상호' : store, '평점' : score, '위치' : place, '메뉴' : menu})
```

```python
line
```

```python
df = pd.DataFrame(line)
df
```

```python
df.to_csv('식신맛집(신논현역).csv', index=False)
```
