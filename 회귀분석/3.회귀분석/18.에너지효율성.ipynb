```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#에너지 효율성 데이터
df = pd.read_csv("c:/data/energy/ENB2012_data.csv")
df.head() 
#X1: 상대적 크기, X2: 건축 표면적, X3: 벽체 면적, X4: 지붕 면적, X5: 전체 높이
#X6: 건물의 방위, X7: 유리창면적, X8: 유리창면적의 분산
#Y1: 난방 하중, Y2: 냉방 하중
```

# 난방효율성

```python
train_cols = df.columns[0:8]
X = df[train_cols]
y = df["Y1"] #난방 하중
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
#다중 회귀분석
import statsmodels.api as sm

model = sm.OLS(y_train, X_train)
result = model.fit()
result.summary()
```

```python
#X6을 제외한 모델
train_cols = df.columns[[0,1,2,3,4,6,7]]
X = df[train_cols]
y = df["Y1"] #난방 하중
```

```python
model = sm.OLS(y, X)
result = model.fit()
result.summary()
# X1(상대적 크기) : 건물이 크면 난방 비용이 감소
# X5(건물 높이) : 건물 높이가 높으면 난방 비용이 증가
# X7(유리창면적) : 유리창면적이 크면 난방 비용이 증가
```

```python
#다중 회귀분석
import statsmodels.api as sm

model = sm.OLS(y_train, X_train)
result = model.fit()
result.summary()
```
