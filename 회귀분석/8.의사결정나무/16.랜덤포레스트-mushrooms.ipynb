```python
import pandas as pd
df = pd.read_csv('c:/data/mushroom/mushrooms.csv')
df.head()
```

```python
df.shape
```

```python
#카테고리형 데이터 > 숫자 변환(정수인코딩)
# 가중치가 붙을 수도 있음.
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
for col in df.columns:
    df[col] = lbl.fit_transform(df[col])
    print(lbl.classes_)
```

```python
df.head()
```

```python
X = df.iloc[:, 1:]
y = df['class']
y.value_counts() # 0: 식용, 1: 독버섯
```

```python
#언더샘플링
from imblearn.under_sampling import RandomUnderSampler
X_sample, y_sample = RandomUnderSampler(random_state=0).fit_resample(X, y)
X_samp = pd.DataFrame(data=X_sample, columns=X.columns)
y_samp = pd.DataFrame(data=y_sample, columns=['class'])

#원핫인코딩
X_samp = pd.get_dummies(X_samp, columns=X_samp.columns, drop_first=True)
# drop_first=True: 원본 데이터 삭제

df_samp = pd.concat([X_samp, y_samp], axis=1)
df_samp.head()
```

```python
df_samp['class'].value_counts()
```

```python
X = df_samp.iloc[:, :-1]
y = df_samp['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
#모형 평가
from sklearn.metrics import accuracy_score, classification_report
pred = model.predict(X_test)
score = accuracy_score(y_test, pred)
report = classification_report(y_test, pred)
print('정답률:', score)
print('리포트:\n', report)
```

```python
#특성중요도
import numpy as np
import matplotlib.pyplot as plt
def plot_tree(model):
    n_features = X.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), X.columns)
    plt.ylim(-1, n_features)

plt.figure(figsize=(20,15))
plot_tree(model)    
```
