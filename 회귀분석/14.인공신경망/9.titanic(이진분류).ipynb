```python
train_cols = df.columns[1: -1]
X = df[train_cols]
y = df['Survived']
```

```python
#언더샘플링
from imblearn.under_sampling import RandomUnderSampler
X_sample, y_sample = RandomUnderSampler(random_state=0).fit_resample(X, y)
X_samp = pd.DataFrame(data=X_sample, columns=train_cols)
y_samp = pd.DataFrame(data=y_sample, columns=['Survived'])
df_samp = pd.concat([X_samp, y_samp], axis=1)
```

```python
X = X_samp
y = y_samp
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    stratify=y, random_state=7)
```

```python
#모델 구성
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(128, input_shape=(len(X_train.columns), ), activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# 모델 학습과정 설정
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model,show_shapes=True)
```

```python
X_train.boxplot()
plt.show()
```

```python
#스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 데이터프레임으로 변경
X_train_scaled = pd.DataFrame(X_train_scaled)
X_test_scaled = pd.DataFrame(X_test_scaled)
```

```python
X_train_scaled.boxplot()
plt.show()
```

```python
#스케일링 모델 저장
from joblib import dump
dump(scaler, 'c:/data/titanic/scaler.model', compress=True)
```

```python
#모델 학습
hist = model.fit(X_train_scaled, y_train, epochs=500)
plt.figure(figsize=(12,8))
plt.plot(hist.history['loss'])
plt.plot(hist.history['accuracy'])
plt.legend(['loss', 'accuracy'])
plt.show()
```

```python
#모델 평가
# 학습용
scores = model.evaluate(X_train_scaled, y_train, verbose=1)
print(scores)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```

```python
# 검증용
scores = model.evaluate(X_test_scaled, y_test, verbose=1)
print(scores)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```
