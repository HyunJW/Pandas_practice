```python
%matplotlib inline
from matplotlib import pyplot as plt
import mglearn
from matplotlib import font_manager, rc
#한글 처리를 위해 폰트 설정
font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
#forge데이터셋 생성
X, y = mglearn.datasets.make_forge()
print(X)
print(y)
```

```python
#산점도 그리기
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(['클래스 0', '클래스 1'], loc=4)
plt.xlabel('첫번째 특성')
plt.ylabel('두번째 특성')
```

```python
#데이터포인트 26개와 특성 2개
print('X.shape: {}'.format(X.shape))
```

```python
#분류용 선형 모델(로지스틱 회귀): 회귀분석이 아닌 분류알고리즘
from sklearn.linear_model import LogisticRegression
fig, axes = plt.subplots(1,1, figsize=(5,3))
model = LogisticRegression()

#모델 학습
model = model.fit(X, y)
mglearn.plots.plot_2d_separator(model, X, fill=True, eps=0.5, ax=axes, alpha=.7)

#산점도 그리기
mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=axes)
axes.set_title('로지스틱 회귀분석')
axes.set_xlabel('특성 0')
axes.set_ylabel('특성 1')
axes.legend()
```

```python
#분류 정확도
model.score(X, y)
```

```python
#정오분류표
from sklearn.metrics import confusion_matrix
pred = model.predict(X)
print(pred)
print(y)
cm = confusion_matrix(y, pred)
print(cm)
```
